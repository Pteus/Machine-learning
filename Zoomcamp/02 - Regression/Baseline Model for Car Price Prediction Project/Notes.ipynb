First, we start with a simple model using only numerical columns.

We choose the columns `engine_hp`, `engine_cylinders`, `highway_mpg`, `city_mpg`, and `popularity` for our base model.

We notice that some of the features in `df_train` contain NaN values. For simplicity, we will set these NaN values to 0 to ensure the model is solvable. However, using a non-zero value, such as the mean value of the feature, would be more appropriate as a filler. Because that means there is an observation of a car with 0 cylinders or 0 horse powers. And a car without cylinders or 0 horse powers does not make much sense at this point. For the current example this procedure is sufficient for us.

The linear regression equation is given by:

$g(x_i) = w_0 + x_{i1}w_1 + x_{i2}w_2$

If $x_{i1} = 0$, then the equation simplifies to:

$g(x_i) = w_0 + 0 + x_{i2}w_2$

Which further reduces to:

$g(x_i) = w_0 + x_{i2}w_2$


Once the weights are calculated, we apply them to derive the predicted $ y $ vector. Then, we plot both the predicted $ y $ and the actual $ y $ on the same histogram for a visual comparison.

```python
sns.histplot(y_pred, color='red', bins = 50, alpha=0.5, label='Predicted y')

````


From this plot, you can see that the model is not ideal. However, it's important to have an objective method to determine whether a model is good or not. As we work on improving the model, we need to ensure that our changes are truly enhancing its performance.
